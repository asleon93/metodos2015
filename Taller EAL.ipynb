```python
import numpy as np
#from openmat import *

print "\nSistema de Ecuaciones:"
print "50 = 5X3 + 2X2"
print "10-X1 = X3"
print "3X2 + 8X1 = 20"

print "\na) Forma Matricial Coeficientes (Matriz Aumentada):"
M = np.array([[0, 2, 5, 50], [-1, 0, -1, -10], [8, 0, 3, 20]])

def transpuesta(matriz):
        rows = len(matriz)
        cols = len(matriz[0])
        return [[matriz[j][i] for j in xrange(rows)] for i in xrange(cols)]
     
print "Original"
print_r(M)
print "\nb) Transpuesta Matriz Coeficientes"
print "TRANSPUESTA"
print_r(transpuesta(M))
```
